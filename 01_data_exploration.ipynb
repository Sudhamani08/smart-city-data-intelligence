```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

air = pd.read_csv("../raw/air_quality.csv")
traffic = pd.read_excel("../data/traffic_data.xlsx")
```

```python
air.columns
traffic.columns
```

```python
air.columns = air.columns.str.strip()
traffic.columns = traffic.columns.str.strip()
```

```python
air.columns
```

```python
traffic.columns
```

```python
traffic['Year'] = traffic['Year'].astype(int)
```

```python
air.columns
```

```python
air.columns
```

```python
air['Date'] = pd.to_datetime(air['Date'])
air['Year'] = air['Date'].dt.year
```

```python
air[['Year']].head()
```

```python
merged_df = pd.merge(
    traffic,
    air,
    on=['City', 'Year'],
    how='inner'
)
```

```python
merged_df.head()
```

```python
merged_df.shape
```

```python
merged_df[['Average_Daily_Traffic_Counts', 'Air_Quality_Index']].corr()
```

```python
merged_df.head()
```

```python
traffic_year = merged_df.groupby('Year')['Average_Daily_Traffic_Counts'].mean()

traffic_year.plot(
    kind='line',
    title='Average Daily Traffic Over Years',
    xlabel='Year',
    ylabel='Traffic Count'
)
```

```python
air_year = merged_df.groupby('Year')['Air_Quality_Index'].mean()

air_year.plot(
    kind='line',
    title='Air Quality Index Over Years',
    xlabel='Year',
    ylabel='AQI'
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
air = pd.read_csv("../raw/air_quality.csv")
traffic = pd.read_excel("../data/traffic_data.xlsx")

# Basic check
air.head(), traffic.head()
```

```python
air.columns, traffic.columns
```

```python
plt.figure()
sns.histplot(air["AQI"], bins=20, kde=True)
plt.title("AQI Distribution")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()
```

```python
air["Date"] = pd.to_datetime(air["Date"])

plt.figure()
sns.lineplot(data=air, x="Date", y="AQI")
plt.title("AQI Trend Over Time")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.show()
```

```python
traffic.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.barplot(
    data=traffic,
    x="City",
    y="Average_Daily_Traffic_Counts",
    estimator=sum,
    errorbar=None
)

plt.title("Total Average Daily Traffic by City")
plt.xlabel("City")
plt.ylabel("Traffic Count")
plt.xticks(rotation=45)
plt.show()
```

```python
traffic.info()
```

```python
traffic.describe()
```

```python
traffic_numeric = traffic.select_dtypes(include="number")
traffic_numeric.head()
```

```python
corr = traffic_numeric.corr()
corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Traffic Metrics")
plt.show()
```

### 🔍 Key Insights from Traffic Data

- Cities with higher population tend to have higher daily traffic counts.
- Average traffic speed decreases as traffic volume increases.
- Number of road accidents shows a positive correlation with traffic volume.
- Peak hourly traffic strongly influences daily congestion levels.
